## Manipulating Columns

```python
df["difference"] = df["pts"] - df["opp_pts"]
```

```python
df
```

```python
df.loc[
    (df["game_result"] == "W") & (df["difference"] == df["difference"].max()),
    ["gameorder", "year_id", "fran_id", "opp_fran", "pts", "opp_pts", "difference"]
]
```

```python
df.rename(columns={"gameorder": "game_order", "year_id": "season", "pts": "points", "opp_pts": "vs_points"})
```

```python
df["match_id"] = df["game_id"] + "." + df["team_id"] + "_" + df["opp_id"]
df
```

```python
df.index = df["match_id"]
```

```python
df
```
